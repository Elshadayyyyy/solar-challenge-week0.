```python
cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Compute Z-scores
from scipy.stats import zscore
z_scores = df[cols_to_check].apply(zscore)

# Flag rows with |Z| > 3
outliers = (np.abs(z_scores) > 3)
outliers_count = outliers.sum()
outliers_count
```

```python
# Fill missing values in key columns with median
for col in cols_to_check:
    df[col] = df[col].fillna(df[col].median())

# Optional: drop extreme outliers (|Z| > 3)
df_cleaned = df[(np.abs(z_scores) <= 3).all(axis=1)]
df_cleaned.shape
```

```python

cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

from scipy.stats import zscore
z_scores = df[cols_to_check].apply(zscore)

outliers = (z_scores.abs() > 3)
print(outliers.sum()) 

df_clean = df[~outliers.any(axis=1)]
print(f"Rows before cleaning: {len(df)}, after cleaning: {len(df_clean)}")
```

```python

# Plot GHI
plt.figure(figsize=(12, 4))
plt.plot(df_clean['Timestamp'], df_clean['GHI'])
plt.title('Global Horizontal Irradiance (GHI) over Time')
plt.xlabel('Timestamp')
plt.ylabel('GHI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot DNI
plt.figure(figsize=(12, 4))
plt.plot(df_clean['Timestamp'], df_clean['DNI'], color='orange')
plt.title('Direct Normal Irradiance (DNI) over Time')
plt.xlabel('Timestamp')
plt.ylabel('DNI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot DHI
plt.figure(figsize=(12, 4))
plt.plot(df_clean['Timestamp'], df_clean['DHI'], color='green')
plt.title('Diffuse Horizontal Irradiance (DHI) over Time')
plt.xlabel('Timestamp')
plt.ylabel('DHI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot Tamb (Ambient Temperature)
plt.figure(figsize=(12, 4))
plt.plot(df_clean['Timestamp'], df_clean['Tamb'], color='red')
plt.title('Ambient Temperature (Tamb) over Time')
plt.xlabel('Timestamp')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by Cleaning flag and calculate average ModA & ModB
cleaning_avg = df.groupby('Cleaning')[['ModA', 'ModB']].mean()
print(cleaning_avg)

# Plot pre/post-cleaning averages
cleaning_avg.plot(kind='bar', figsize=(8,5), color=['skyblue', 'orange'])
plt.title('Average Module Readings Pre/Post Cleaning')
plt.xlabel('Cleaning (0 = Before, 1 = After)')
plt.ylabel('Average Module Reading')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python


# Correlation heatmap
corr_cols = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB']
plt.figure(figsize=(8,6))
sns.heatmap(df[corr_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Scatter plots
plt.figure(figsize=(6,4))
plt.scatter(df['WS'], df['GHI'], alpha=0.5)
plt.title('Wind Speed vs GHI')
plt.xlabel('WS (m/s)')
plt.ylabel('GHI')
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df['WSgust'], df['GHI'], alpha=0.5, color='orange')
plt.title('Wind Gust vs GHI')
plt.xlabel('WSgust (m/s)')
plt.ylabel('GHI')
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df['RH'], df['Tamb'], alpha=0.5, color='green')
plt.title('Relative Humidity vs Ambient Temperature')
plt.xlabel('RH (%)')
plt.ylabel('Tamb (C)')
plt.show()
```

```python
# Histogram for GHI
plt.figure(figsize=(8,4))
plt.hist(df['GHI'], bins=50, color='gold', edgecolor='black')
plt.title('Distribution of GHI')
plt.xlabel('GHI')
plt.ylabel('Frequency')
plt.show()

# Histogram for Wind Speed (WS)
plt.figure(figsize=(8,4))
plt.hist(df['WS'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Wind Speed')
plt.xlabel('WS (m/s)')
plt.ylabel('Frequency')
plt.show()

# Wind Rose
# Convert wind direction to radians
theta = np.deg2rad(df['WD'])
r = df['WS']

plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
ax.scatter(theta, r, alpha=0.3)
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
plt.title('Wind Rose (WS vs WD)')
plt.show()
```

```python
# Scatter plot: RH vs Tamb
plt.figure(figsize=(8,5))
plt.scatter(df['RH'], df['Tamb'], alpha=0.3, color='orange')
plt.title('Relative Humidity vs Ambient Temperature')
plt.xlabel('RH (%)')
plt.ylabel('Tamb (C)')
plt.show()

# Scatter plot: RH vs GHI
plt.figure(figsize=(8,5))
plt.scatter(df['RH'], df['GHI'], alpha=0.3, color='green')
plt.title('Relative Humidity vs Global Horizontal Irradiance (GHI)')
plt.xlabel('RH (%)')
plt.ylabel('GHI')
plt.show()
```
